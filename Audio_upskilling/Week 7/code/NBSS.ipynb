```python
import torchaudio
import librosa
# from source_separation import NBSS
from IPython.display import Audio, display
```

```python
waveform, sample_rate = torchaudio.load('/content/mixture_mic1.wav', normalize=True)

# Instantiate the model (e.g., with NB_BLSTM architecture)
model = NBSS(n_channel=1, n_speaker=2, arch="NB_BLSTM")  # Assuming mono audio, adjust n_channel if needed

# Forward pass to obtain separated audio signals
ys_hat = model(waveform.unsqueeze(0))  # Add batch dimension

# Convert the tensor to a NumPy array
separated_audio = ys_hat.squeeze().detach().numpy()
```

```python
display(Audio("/content/mixture_mic1.wav", rate=sample_rate))
display(Audio(separated_audio[0], rate=sample_rate))
display(Audio(separated_audio[1], rate=sample_rate))
```
