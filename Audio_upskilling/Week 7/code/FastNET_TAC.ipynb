```python
import torch
import librosa
from FaSNet import FaSNet_TAC

# Step 1: Load an Audio File
input_audio, sr = librosa.load('/content/mixture_mic1.wav', sr=16000)  # Change the path to your audio file

```

```python
model = FaSNet_TAC(enc_dim=64, feature_dim=64, hidden_dim=128, layer=4)
```

```python
# Step 4: Prepare Input Data
input_tensor = torch.tensor([input_audio,input_audio,input_audio]).unsqueeze(0)  # Add batch dimension

```

```python
input_tensor.shape
```

```python
# Step 5: Perform Source Separation
num_mic = torch.tensor([3])  # Assuming n_channel is the number of channels in your input
separated_audio = model(input_tensor, num_mic)

```

```python
# Step 6: Postprocess the Separated Audio (if needed)

# Optional: Convert tensor to numpy array for audio playback
separated_audio_np = separated_audio.squeeze().cpu().detach().numpy()
```

```python
separated_audio_np.shape
```

```python
from IPython.display import Audio, display

display(Audio("/content/mixture_mic1.wav",rate=sr))
display(Audio(separated_audio_np[0],rate=sr))
display(Audio(separated_audio_np[1],rate=sr))
```
